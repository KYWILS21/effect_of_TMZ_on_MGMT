```python
import pandas as pd

meth_data = pd.read_csv('TCGA_data/cleaned_methylation.csv', index_col=0)
clinical_df = pd.read_csv('TCGA_data/clean_phenotype_data.csv')

meth_data.index = meth_data.index.str[:12]
clinical_df['Matching_ID'] = clinical_df['sample'].str[:12]

meth_collapsed = meth_data.groupby(meth_data.index).mean()

probe_id = 'cg00000029'
analysis_subset = meth_collapsed[[probe_id]].copy()
analysis_subset['Methylation_Status'] = (analysis_subset[probe_id] >= 0.3).astype(float)


final_df = pd.merge(
    clinical_df,
    analysis_subset,
    left_on='Matching_ID',
    right_index=True,
    how='left'
)

final_df.to_csv('TCGA_data/survival_full_set.csv', index=False)

print(f"Total clinical patients kept: {len(final_df)}")
print(f"Patients with missing methylation data: {final_df[probe_id].isna().sum()}")
```
