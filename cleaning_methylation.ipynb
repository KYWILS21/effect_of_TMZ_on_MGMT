```python
import pandas as pd
import numpy as np

file_path = 'TCGA_data/DNA_methylation.txt'


df = pd.read_csv(file_path, sep='\t', index_col=0, low_memory=False)

df = df.apply(pd.to_numeric, errors='coerce').astype(np.float32)

required_completeness = 0.80
limit = int(required_completeness * df.shape[1])

df_filtered = df.dropna(thresh=limit, axis=0)

print(f"Original probes: {df.shape[0]}")
print(f"Probes kept (80% completeness required): {df_filtered.shape[0]}")

df_final = df_filtered.T

df_final.to_csv('TCGA_data/cleaned_methylation.csv')
```
